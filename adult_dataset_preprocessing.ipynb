```
import pandas as pd
import numpy as np
import sklearn
missing_values = ["n/a", "na", "--", " ", "  ", "   ", "    ", "? "]
column_names = ["age", "workclass" , "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race" , "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country", "Target"]
df = pd.read_csv('/content/adult.csv', na_values = missing_values, names=column_names)
df.head()
```

```
df_target = df['Target']
df_target = np.array(df_target)
i = -1
for value in df_target:
  i += 1
  if "<=50K" in value:
    df_target[i] = "0"
  if ">50K" in value:
    df_target[i] = "1"
    
```

```
df_target = pd.DataFrame(df_target)
```

```
df['Target'] = df_target
```

```
df.tail()
```

```
age_col = df['age']
age_col = np.array(age_col)
age_col = age_col.reshape(-1, 1)
age_col.shape
```

```
from sklearn.preprocessing import KBinsDiscretizer
enc = KBinsDiscretizer(n_bins=5, encode='ordinal')
age_binned = enc.fit_transform(age_col)
```

```
age = pd.DataFrame(age_binned)
df['age'] = age 
```

```
df = df.drop(['fnlwgt'], axis=1)
```

```
df = df.drop(['education-num'], axis=1)
```

```
df.head()
```

```
hrs_col = df['hours-per-week']
hrs_col = np.array(hrs_col)
hrs_col = hrs_col.reshape(-1, 1)
enc = KBinsDiscretizer(n_bins=5, encode='ordinal')
hrs_binned = enc.fit_transform(hrs_col)
hrs = pd.DataFrame(hrs_binned)
df['hours-per-week'] = hrs
```

```
df.tail()
```

```
cap_gain_list = []
cap_gain = np.array(df['capital-gain'])
for value in cap_gain:
  value = int(value)
  if value > 0:
    cap_gain_list.append(value)
cap_gain_list.sort()
```

```
print(cap_gain_list)
```

```
import statistics
cap_gain_median = statistics.median(cap_gain_list)
```

```
i = -1
for value in cap_gain:
  i += 1
  value = int(value)
  if value == 0:
    continue
  elif value > 0 and value <= cap_gain_median:
    cap_gain[i] = "1"
  elif value > cap_gain_median:
    cap_gain[i] = "2"

```

```
print(cap_gain)
```

```
capital_gain = pd.DataFrame(cap_gain)
df['capital-gain'] = capital_gain
df.head()
```

```
cap_loss_list = []
cap_loss = np.array(df['capital-loss'])
for value in cap_loss:
  value = int(value)
  if value > 0:
    cap_loss_list.append(value)

cap_loss_median = statistics.median(cap_loss_list)
i = -1
for value in cap_loss:
  i += 1
  value = int(value)
  if value == 0:
    continue
  elif value > 0 and value <= cap_loss_median:
    cap_loss[i] = "1"
  elif value > cap_loss_median:
    cap_loss[i] = "2"

capital_loss = pd.DataFrame(cap_loss)
df['capital-loss'] = capital_loss
df.head(50)
```

```
df.to_csv ('/content/adult_preprocessed.csv', index = False, header=True)
```
