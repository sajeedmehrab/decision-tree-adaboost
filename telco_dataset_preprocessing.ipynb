```
import pandas as pd
import numpy as np
import sklearn
missing_values = ["n/a", "na", "--", " ", "  ", "   ", "    "]
df = pd.read_csv('/content/WA_Fn-UseC_-Telco-Customer-Churn.csv', na_values = missing_values)
df.head()
```

```
df.drop(['customerID'], axis=1, inplace=True)
```

```
df.head()
```

```
type(df)
```

```
df.Churn.replace(('Yes', 'No'), (1, 0), inplace=True)
```

```
print(df.isnull().sum())
```

```
df['TotalCharges'].median()
```

```
df['TotalCharges'].mean()
```

```
mean_totalcharges = df['TotalCharges'].mean()
df['TotalCharges'].fillna(mean_totalcharges, inplace=True)
```

```
print(df.isnull().sum())
```

```
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler() 

monthly_charges = (np.array(df['MonthlyCharges'])).reshape(-1, 1)
```

```
monthly_charges.shape
```

```
monthly_charges = scaler.fit_transform(monthly_charges)
monthly_charges = pd.DataFrame(monthly_charges)
monthly_charges.head()
```

```
df['MonthlyCharges'] = monthly_charges
df.head()
```

```
total_charges = (np.array(df['TotalCharges'])).reshape(-1, 1)
total_charges = scaler.fit_transform(total_charges)
total_charges = pd.DataFrame(total_charges)
df['TotalCharges'] = total_charges
df.head()
```

```
from sklearn.preprocessing import KBinsDiscretizer
enc = KBinsDiscretizer(n_bins=20, encode='ordinal')
tenure = (np.array(df['tenure'])).reshape(-1, 1)
tenure = enc.fit_transform(tenure)
tenure = pd.DataFrame(tenure)
df['tenure'] = tenure
df.head()
```

```
TotalCharges = (np.array(df['TotalCharges'])).reshape(-1, 1)
TotalCharges = enc.fit_transform(TotalCharges)
TotalCharges = pd.DataFrame(TotalCharges)
df['TotalCharges'] = TotalCharges
df.head()
```

```
MonthlyCharges = (np.array(df['MonthlyCharges'])).reshape(-1, 1)
MonthlyCharges = enc.fit_transform(MonthlyCharges)
MonthlyCharges = pd.DataFrame(MonthlyCharges)
df['MonthlyCharges'] = MonthlyCharges

df.rename(columns = {'Churn':'Target'}, inplace = True) 
df.head()
```

```
df.to_csv ('/content/telco_preprocessed.csv', index = False, header=True)
```
